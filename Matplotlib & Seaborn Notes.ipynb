```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from numpy.random import randn, randint, rand
np.random.seed(101)
%matplotlib inline # required for jupyter notebook
```

```python
x = np.linspace(0,5,11)
y = x ** 2
plt.plot(x, y) # functional method
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
# plt.show() if not in jupyter notebook
```

```python
plt.subplot(1, 2, 1) # num_rows, num_cols, plot_num refering to
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
 # OO approach
fig = plt.figure()
axes = fig.add_axes([.1, .1, .8, .8]) # % from left , % from bot, width % of canvis size, height % of canvis size
axes.plot(x, y)
axes.set_xlabel('X label')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1, .1, .8, .8])
axes2 = fig.add_axes([.2, .5, .4, .3])
axes1.plot(x, y)
axes2.plot(y, x)
axes1.set_title('Bigger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x, y)
axes[0].set_title('Plot 0')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(3,2)) # 3 inches wide by 2 inches tall
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(3,4))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('my_plot.png', dpi=200) # save picture of plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='x squared')
ax.plot(x, x ** 3, label='x cubed')
ax.legend() # loc.legend(loc=0) # 0: best location, 1: upper right, 2: upper left, 3: low lt, 4: lw rt, 10: center
# loc.legend(loc=(.1, .1))
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='green', lw=2, alpha=.5, linestyle='--', marker='o', markersize=10) # or can use RGB hex code: color='#FF8C00'
# linewidth: lw=1 is default, alpha (transparency from 0 to 1)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='green')
ax.set_xlim([0,1]) # define boundary for x axis
ax.set_ylim([0,5])
```
