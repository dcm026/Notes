```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                   'B': ['B0', 'B1', 'B2', 'B3'],
                   'C': ['C0', 'C1', 'C2', 'C3'],
                   'D': ['D0', 'D1', 'D2', 'D3']},
                  index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                   'B': ['B4', 'B5', 'B6', 'B7'],
                   'C': ['C4', 'C5', 'C6', 'C7'],
                   'D': ['D4', 'D5', 'D6', 'D7']},
                  index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                   'B': ['B8', 'B9', 'B10', 'B11'],
                   'C': ['C8', 'C9', 'C10', 'C11'],
                   'D': ['D8', 'D9', 'D10', 'D11']},
                  index=[8, 9, 10, 11])
```

```python
df1
```

```python
pd.concat([df1,df2,df3]) # default merge on rows (axis=0)
```

```python
pd.concat([df1,df2,df3], axis=1) # merge on columns
```

```python
left = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})
right = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                    'C': ['c0', 'c1', 'c2'],
                    'D': ['d0', 'd1', 'd2']})
left
```

```python
right
```

```python
pd.merge(left, right, how='inner', on='key')
```

```python
left = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                     'key2': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2']})
right = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                      'key2': ['k0', 'k1', 'k2'],
                    'C': ['c0', 'c1', 'c2'],
                    'D': ['d0', 'd1', 'd2']})
```

```python
pd.merge(left, right, how='inner', on=['key', 'key2'])
```

```python
# how many people have the word chief in JobTitle?
def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
sum(df['JobTitle'].apply(lambda x: chief_string(x)))
```

```python
'''MERGING/JOINING REVISTED'''
```

```python
df1 = pd.DataFrame({'id': ['0', '1'],
                    'B': ['10', '11']})
df2 = pd.DataFrame({'id': ['1', '2'],
                    'C': ['0', np.nan]})
df1
```

```python
df2
```

```python
print('LEFT JOIN on id')
pd.merge(df1, df2, on='id', how='left')
```

```python
print('INNER JOIN on id')
pd.merge(df1, df2, on='id', how='inner')
```

```python
print('RIGHT JOIN on id')
pd.merge(df1, df2, on='id', how='right')
```

```python
print('OUTER JOIN on id')
pd.merge(df1, df2, on='id', how='outer') # (inner and outer)
```
